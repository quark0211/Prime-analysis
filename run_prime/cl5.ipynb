```julia
using Pkg
Pkg.activate(".")
# using Prime
```

```julia
using Distributed
addprocs(60)
@everywhere using Prime
```

```julia
using HDF5,DelimitedFiles,StatsBase
#Calculate cell-volume factor β
f = h5open("dataset/real_data/cl5_1193all.loom", "r")
X = read(f["layers/spliced"])+read(f["layers/unspliced"])
total_rd=sum(X,dims=2)
β = total_rd./mean(total_rd)

rdm = Matrix(readdlm("dataset/real_data/spliced_expression_matrix_5cl.csv",',')[2:end,2:end]')
rdn = Matrix(readdlm("dataset/real_data/unspliced_expression_matrix_5cl.csv", ',')[2:end,2:end]')

#cell number and gene number
n,p = size(rdm)
#Define cluster number and select model
k = 5
sel = 4
```

```julia
cluster_labels, γ = Prime.prime_cluster(
    rdn, rdm, β, k;
    sel = sel,
    maxiter = 20,  
)
```

```julia
using Clustering
cell_type = readdlm("dataset/real_data/subclass_labels_5cl.csv",',')[2:end,2]
vec_numeric = replace(cell_type, "HCC827" => 1, "H1975" => 2, "H2228" => 4, "A549" => 3, "H838" => 5)

println("ARI for PRIME clustering:",randindex(Int.(cluster_labels),Int.(vec_numeric))[1])
println("NMI for PRIME clustering:",mutualinfo(Int.(cluster_labels),Int.(vec_numeric)))
```
