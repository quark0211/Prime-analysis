```julia
using HDF5,DelimitedFiles,StatsBase
#Calculate cell-volume factor β
f = h5open("dataset/real_data/allen_b08_1948all.loom", "r")
X = read(f["layers/spliced"])+read(f["layers/unspliced"])
total_rd=sum(X,dims=2)
β = total_rd./mean(total_rd)

rdm = Matrix(readdlm("dataset/real_data/spliced_expression_matrix_Mopsc_682.csv",',')[2:end,2:end]')
rdn = Matrix(readdlm("dataset/real_data/unspliced_expression_matrix_Mopsc_682.csv", ',')[2:end,2:end]')

#cell number and gene number
n,p = size(rdm)
#Define cluster number and select model
k = 10
sel = 4
```

```julia
cluster_labels, γ = Prime.prime_cluster(
    rdn, rdm, β, k;
    sel = sel,
    maxiter = 20,  
)
```

```julia
using Clustering
cell_type = readdlm("dataset/real_data/subclass_labels_Mopsc_682.csv",',')[2:end,2]
vec_numeric = replace(cell_type, 
"L6 IT"   => 1,
"L6 CT"   => 2,
"L5 IT"   => 3,
"Lamp5"   => 4,
"Vip"     => 5,
"Sst"     => 6,
"L2/3 IT" => 7,
"L5/6 NP" => 8,
"L6b"     => 9,
"Pvalb"   => 10)
println("ARI for PRIME clustering:",randindex(Int.(cluster_labels),Int.(vec_numeric))[1])
println("NMI for PRIME clustering:",mutualinfo(Int.(cluster_labels),Int.(vec_numeric)))
```
